```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Define initial variables
monthly_inflation = 0.033  # 3.3% per month
yearly_inflation = 0.248   # 24.8% per year
nominal_salary = 70000  # Salary per month

# Compute Real Salary (adjusted for inflation)
real_salary_monthly = nominal_salary / (1 + monthly_inflation)
real_salary_yearly = (nominal_salary * 12) / (1 + yearly_inflation)

# Assume labor productivity is proportional to real wages (simplified model)
# Define a base productivity level
base_productivity = 100  # Arbitrary starting value

# Productivity decreases as inflation rises (inverse relationship assumption)
productivity_monthly = base_productivity * (real_salary_monthly / nominal_salary)
productivity_yearly = base_productivity * (real_salary_yearly / (nominal_salary * 12))

# Create DataFrame for visualization
data = {
    "Period": ["Monthly", "Yearly"],
    "Nominal Salary": [nominal_salary, nominal_salary * 12],
    "Real Salary": [real_salary_monthly, real_salary_yearly],
    "Labor Productivity": [productivity_monthly, productivity_yearly],
    "Inflation Rate": [monthly_inflation * 100, yearly_inflation * 100]
}

df = pd.DataFrame(data)

# Print the calculated values
print("Data Analysis:")
print(df)

# Plot Inflation vs. Productivity
plt.figure(figsize=(8, 5))
plt.bar(df["Period"], df["Labor Productivity"], color=['blue', 'red'])
plt.xlabel("Time Period")
plt.ylabel("Labor Productivity Index")
plt.title("Impact of Inflation on Labor Productivity")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Constants
minimum_wage = 70000  # Annual minimum wage in Naira
initial_inflation_rate = 34.8 / 100  # Initial inflation rate
reduced_inflation_rate = 18.2 / 100  # Reduced inflation rate

# Function to calculate the effective wage after inflation
def effective_wage(minimum_wage, inflation_rate):
    return minimum_wage / (1 + inflation_rate)

# Calculate effective wages
effective_wage_initial = effective_wage(minimum_wage, initial_inflation_rate)
effective_wage_reduced = effective_wage(minimum_wage, reduced_inflation_rate)

# Monthly and Yearly calculations
monthly_initial = effective_wage_initial / 12
monthly_reduced = effective_wage_reduced / 12

yearly_initial = effective_wage_initial
yearly_reduced = effective_wage_reduced

# Output results
print("Impact of Inflation on Productivity")
print("=======================================")
print(f"Initial Inflation Rate: {initial_inflation_rate * 100}%")
print(f"Reduced Inflation Rate: {reduced_inflation_rate * 100}%")
print("\nEffective Wage Analysis:")
print(f"Effective Yearly Wage (Initial): ₦{yearly_initial:.2f}")
print(f"Effective Yearly Wage (Reduced): ₦{yearly_reduced:.2f}")
print(f"Effective Monthly Wage (Initial): ₦{monthly_initial:.2f}")
print(f"Effective Monthly Wage (Reduced): ₦{monthly_reduced:.2f}")

# Productivity Impact
productivity_increase = ((effective_wage_reduced - effective_wage_initial) / effective_wage_initial) * 100
print(f"\nProductivity Increase Due to Inflation Reduction: {productivity_increase:.2f}%")

# Plotting the results
labels = ['Initial Effective Yearly Wage', 'Reduced Effective Yearly Wage']
yearly_wages = [yearly_initial, yearly_reduced]

plt.figure(figsize=(10, 6))
plt.bar(labels, yearly_wages, color=['red', 'green'])
plt.ylabel('Effective Wage (₦)')
plt.title('Impact of Inflation on Effective Yearly Wage')
plt.ylim(0, max(yearly_wages) * 1.2)  # Adding some space above the bars
for i, v in enumerate(yearly_wages):
    plt.text(i, v + 1000, f'₦{v:.2f}', ha='center', va='bottom')

plt.grid(axis='y')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Inflation rates from January to May 2025
inflation_rates = [24.8, 23.18, 24.23, 23.71, 22.97]

# Minimum wage
minimum_wage = 70000  # in naira

# Calculate the impact of inflation on the minimum wage
adjusted_wages = [minimum_wage * (1 + rate/100) for rate in inflation_rates]

# Months
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']

# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(months, adjusted_wages, marker='o', color='b', label='Adjusted Wage')
plt.axhline(y=minimum_wage, color='r', linestyle='--', label='Minimum Wage')

plt.title('Impact of Inflation on Minimum Wage (Jan-May 2025)')
plt.xlabel('Month')
plt.ylabel('Wage (Naira)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Inflation rates from January to May 2025
inflation_rates = [24.8, 23.18, 24.23, 23.71, 22.97]

# Minimum wages in Lagos and Rivers
lagos_minimum_wage = 100000
rivers_minimum_wage = 85000

# Months
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']

# Function to calculate adjusted wages based on inflation
def calculate_adjusted_wages(minimum_wage, inflation_rates):
    adjusted_wages = [minimum_wage * (1 + rate / 100) for rate in inflation_rates]
    return adjusted_wages

# Calculate adjusted wages for Lagos and Rivers
lagos_adjusted_wages = calculate_adjusted_wages(lagos_minimum_wage, inflation_rates)
rivers_adjusted_wages = calculate_adjusted_wages(rivers_minimum_wage, inflation_rates)

# Plotting the data
plt.figure(figsize=(12, 7))

# Plot Lagos data
plt.plot(months, lagos_adjusted_wages, marker='o', color='blue', label='Lagos Adjusted Wage')
plt.axhline(y=lagos_minimum_wage, color='blue', linestyle='--', label='Lagos Minimum Wage')

# Plot Rivers data
plt.plot(months, rivers_adjusted_wages, marker='o', color='green', label='Rivers Adjusted Wage')
plt.axhline(y=rivers_minimum_wage, color='green', linestyle='--', label='Rivers Minimum Wage')

plt.title('Impact of Inflation on Minimum Wage (Jan-May 2025) - Lagos vs. Rivers')
plt.xlabel('Month')
plt.ylabel('Wage (Naira)')
plt.legend()
plt.grid(True)
plt.show()
```
